```python
import os
from google.colab import userdata
os.environ['GROQ_API_KEY']=userdata.get('GROQ_API_KEY')
```

```python
# import the libraries
from langchain.document_loaders import UnstructuredFileLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_chroma import Chroma
from langchain_groq import ChatGroq
from langchain.chains import RetrievalQA
```

```python
# loading the document
loader = UnstructuredFileLoader('/content/BodoiandRobert.pdf')
documents = loader.load()
```

```python
# splitting into chunks
text_splitter = CharacterTextSplitter(
    chunk_size = 2000,
    chunk_overlap = 400
)
```

```python
texts = text_splitter.split_documents(documents)
```

```python
print(texts)
```

```python
texts[:3]
```

```python
# embedding model
embedding = HuggingFaceEmbeddings()
```

```python
persist_directory = "doc_db"
```

```python
# vector DB
vectordb = Chroma.from_documents(
    documents = texts,
    embedding = embedding,
    persist_directory=persist_directory
)
```

```python
# retriever
retriever = vectordb.as_retriever()
```

```python
# llm from groq
llm = ChatGroq(
    model = "llama-3.1-70b-versatile",
    temperature = 0
)
```

```python
# create a QA chain
qa_chain = RetrievalQA.from_chain_type(
    llm = llm,
    chain_type = "stuff",
    retriever = retriever,
    return_source_documents = True
)
```

```python
# invoke the chain and get a response for user query
query = "What is the model discussed in this paper?"
response = qa_chain.invoke({"query":query})
```

```python
print(response['result'])
```

```python
print(response["source_documents"][0].metadata["source"])
```

```python
# compleate RAG
query = "What is the model discussed in this paper?"
response = qa_chain.invoke({"query":query})
print(response['result'])
print("*"*30)
print("Source Document",response["source_documents"][0].metadata["source"])
```
